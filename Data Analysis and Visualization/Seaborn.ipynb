```python
sns.jointplot(data=df,x='math score',y='reading score')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hist')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',shade=True)
```

```python
sns.jointplot(data=df,x='math score',y='reading score',hue='gender')
```

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='gender')
```

```python
sns.pairplot(data=df,hue='gender',corner=True)
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='lunch')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='test preparation course')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='violin',col='lunch',row='race/ethnicity')
```

```python
g = sns.PairGrid(df,hue='gender')

g = g.map_upper(sns.scatterplot)
```

```python
g = sns.PairGrid(df,hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(df,hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```
